```python
numero_dia = int(input("Digite um número de 1 a 7: "))

if numero_dia == 1:
  print("Domingo")
elif numero_dia == 2:
  print("Segunda-feira")
elif numero_dia == 3:
  print("Terça-feira")
elif numero_dia == 4:
  print("Quarta-feira")
elif numero_dia == 5:
  print("Quinta-feira")
elif numero_dia == 6:
  print("Sexta-feira")
elif numero_dia == 7:
  print("Sábado")
else:
  print("Número inválido. Digite um número de 1 a 7.")
```

```python
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_por_hora = float(input("Digite o valor por hora: "))

if horas_trabalhadas <= 40:
  salario = horas_trabalhadas * valor_por_hora
  print(f"O salário é de R${salario:.2f}")
else:
  horas_normais = 40
  horas_extras = horas_trabalhadas - horas_normais
  salario_normal = horas_normais * valor_por_hora
  salario_extra = horas_extras * valor_por_hora * 1.5
  salario_total = salario_normal + salario_extra
  print(f"O salário é de R${salario_total:.2f}")
```

## for in range

repete uma estrutrura de codigo N vezes que sera determinado por um contador(x,i, ou outra variavel).
Ex:

```python
cont = 0
for cont in range(1001):
  print ("contador" , cont)
```

```python
numero = int(input("digite um numero: "))
for i in range(11):
  resultado = numero * i
  print(f"{numero} x {i} = {resultado}")
```
