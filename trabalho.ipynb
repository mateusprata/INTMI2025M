##OPERADOES

SOMA +

SUBTRAÇÃO -

DIVISÃO /

MULTIPLICAÇÃO *

POTENCIA ^

RESTO DA DIVISÃO %

```python
x = int(input("digite um numero: "))
y = int(input("digite um outro numero: "))

print(f"Sua soma é {x + y}")
print(f"Sua Subtração é {x - y}")
print(f"Sua Divisão é {x / y}")
print(f"Sua Mltiplicação é {x * y}")
```
